```python
import numpy as np
import pandas as pd
import torch

data_raw = pd.read_csv('../data/titanic/raw/train.csv')
```

```python
cat_cols = ['PassengerId', 'Pclass', 'Sex', 'Ticket', 'Cabin', 'Embarked']
data_raw[cat_cols].head()
```

```python
data_cat_col_str = data_raw[cat_cols].applymap(str)
```

```python
from collections import Counter

index_dict = {column: {pair[0]: i for i, pair in enumerate(Counter(data_cat_col_str[column].tolist()).most_common())} for column in data_cat_col_str.columns}

one_hot_data = {}

for column in data_cat_col_str.columns:
    temp_data = np.zeros((len(index_dict[column]), len(data_cat_col_str)))
    for x in data_cat_col_str[column]:
        idx = index_dict[column][x]
        temp_data[idx] = 1
        
    one_hot_data[column] = temp_data
```
